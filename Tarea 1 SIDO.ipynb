```python

#problema 4
def missingNumbers(arr,brr):
  arr_dict={}
  brr_dict={}
  for num in arr:
    if num in arr_dict:
      arr_dict[num]+=1
    else:
      arr_dict[num]=1
    for num in brr:
      if num in brr_dict:
        brr_dict[num] +=1
      else:
        brr_dict[num]=1
    missing_nums=[]
    for num, freq in brr_dict.items():
      if num not in arr_dict:
        missing_nums.append(num)
      elif freq > arr_dict[ num]:
        missing_nums.append(num)
    missing_nums.sort()
    return missing_nums
```

```python
#problema 9
def gameOfStones(n):
  cases={1:False, 2:True, 3:True, 4:True, 5:True}
  def call(n):
    if n not in cases:
      cases[n]= not call(n-2) or not call(n-3) or not call(n-5)
    return cases[n]
  return "First" if call(n) else "Second"
```
